```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.font_manager import FontProperties
import matplotlib as mpl
import numpy as np

# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 正确显示负号

# 创建图形
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')  # 隐藏坐标轴

# 定义节点位置和文本 - 优化布局，避免箭头过长
nodes = {
    'A': {'pos': (2, 8), 'text': '河北屯田'},
    'B': {'pos': (5, 8), 'text': '年收粟米200万石'},  # 缩短水平距离
    'C': {'pos': (2, 6), 'text': '营州互市'},
    'D': {'pos': (5, 6), 'text': '年利150万贯'},     # 缩短水平距离
    'E': {'pos': (2, 4), 'text': '铁冶专营'},
    'F': {'pos': (5, 4), 'text': '兵器自产率100%'}, # 缩短水平距离
    'G': {'pos': (2, 2), 'text': '胡商税'},
    'H': {'pos': (5, 2), 'text': '年入80万贯'}      # 添加新节点
}

# 绘制节点
for key, node in nodes.items():
    x, y = node['pos']
    text = node['text']
    
    # 绘制矩形节点
    rect = patches.Rectangle((x-1.2, y-0.4), 2.4, 0.8, 
                             linewidth=1.5, edgecolor='#1f77b4', 
                             facecolor='#e6f2ff', alpha=0.9,
                             zorder=3)  # 确保节点在箭头上方
    ax.add_patch(rect)
    
    # 添加文本
    plt.text(x, y, text, ha='center', va='center', fontsize=12, zorder=4)

# 绘制箭头连接 - 优化箭头长度
arrow_props = dict(
    arrowstyle='->', 
    lw=2, 
    color='#2ca02c', 
    connectionstyle='arc3,rad=0',
    shrinkA=15,  # 缩短箭头起点
    shrinkB=15   # 缩短箭头终点
)

# 连接各组节点
connections = [
    ('A', 'B'),
    ('C', 'D'),
    ('E', 'F'),
    ('G', 'H')  # 添加新连接
]

for start, end in connections:
    start_pos = nodes[start]['pos']
    end_pos = nodes[end]['pos']
    
    # 绘制箭头
    ax.annotate('', 
                xy=end_pos, 
                xytext=start_pos, 
                arrowprops=arrow_props,
                zorder=2)  # 确保箭头在节点下方

# 添加标题
plt.title('唐代经济政策成效', fontsize=16, pad=20)

# 添加说明文本
plt.text(8, 8.5, '数据来源：天宝十四载(755)财政报告', 
         fontsize=10, color='gray', ha='right')

# 调整布局
plt.tight_layout()
plt.subplots_adjust(top=0.9)  # 为标题留出空间
plt.show()
```

```python
# 时间线：安史之乱前兆
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
import matplotlib as mpl

# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 创建图形
fig, ax = plt.subplots(figsize=(12, 6))

# 隐藏坐标轴
ax.axis('off')

# 时间线数据
events = [
    {'year': 734, 'title': '李林甫颁《长行旨》', 'detail': '中央丧失税基审核权'},
    {'year': 747, 'title': '杨国忠征鲜于仲通讨南诏', 'detail': '二十万大军仅归2万人'},
    {'year': 755, 'title': '安禄山范阳起兵', 'detail': '叛军含粟特武装商团'},
    {'year': 781, 'title': '实行两税法', 'detail': '从按丁征税到按资征税'}
]

# 设置时间范围
start_year = 730
end_year = 790

# 绘制时间线
ax.set_xlim(start_year, end_year)
ax.set_ylim(0, 10)

# 绘制时间轴
ax.hlines(y=5, xmin=start_year, xmax=end_year, color='#1f77b4', linewidth=2)

# 添加事件标记
for i, event in enumerate(events):
    year = event['year']
    title = event['title']
    detail = event['detail']
    
    # 绘制标记点
    ax.plot(year, 5, 'o', markersize=10, color='#d62728')
    
    # 添加垂直线
    ax.vlines(x=year, ymin=5, ymax=7.5, color='#2ca02c', linestyle='--', alpha=0.7)
    
    # 添加事件标题
    ax.text(year, 7.8, title, ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    # 添加事件详情
    ax.text(year, 7.2, detail, ha='center', va='top', fontsize=10, color='#555555')
    
    # 添加年份标签
    ax.text(year, 4.2, f"{year}年", ha='center', va='top', fontsize=10)

# 添加标题
plt.title('肿瘤生长到转移全过程：安史之乱前兆', fontsize=16, pad=20)

# 添加时间线标签
ax.text(start_year-1, 5, '时间线 →', ha='right', va='center', fontsize=10)

# 调整布局
plt.tight_layout()
plt.show()
```
